```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data, target = load_iris(return_X_y=True, as_frame=True)
sl = data['sepal length (cm)']
plt.plot(sl)
plt.xlabel('Índice da Amostra')
plt.ylabel('Comprimento da Sépala')
plt.title('Comprimento da Sépala em Todas as Amostras')
plt.show()
```

```python
sw = data['sepal width (cm)']
plt.plot(sw)
plt.xlabel('Índice da Amostra')
plt.ylabel('Largura da Sépala')
plt.title('Largura da Sépala em Todas as Amostras')
plt.show()
```

```python
pl = data['petal length (cm)']
plt.plot(pl)
plt.xlabel('Índice da Amostra')
plt.ylabel('Comprimento da Pétala')
plt.title('Comprimento da Pétala em Todas as Amostras')
plt.show()
```

```python
pw = data['petal width (cm)']
plt.plot(pw)
plt.xlabel('Índice da Amostra')
plt.ylabel('Largura da Pétala')
plt.title('Largura da Pétala em Todas as Amostras')
plt.show()
```
